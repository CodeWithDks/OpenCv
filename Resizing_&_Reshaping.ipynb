```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
upload = files.upload()
```

```python
filename = next(iter(upload))
```

```python
print(filename)
```

```python
image = cv2.imdecode(np.frombuffer(upload[filename], np.uint8), cv2.IMREAD_COLOR)
```

```python
if image is not None:
  print("image is loaded successfully")
  resized = cv2.resize(image,(72,1280))
  print("Here is your orignal image")
  cv2_imshow(image)
  print(image.shape)
  print('Here is your resize image')
  cv2_imshow(resized)
else:
  print("image not found")
```
